```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

Just some markdown before a start

starting with creation of figure and coordinats

```python
fig = plt.figure()
ax = plt.axes()
```

now can use ax.plot to create graph

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
```

to create graph with multiple lines function plt can be called multiple times

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x-0), color='blue') #  
plt.plot(x, np.sin(x-1), color='g')    #  (rgbcmyk)
plt.plot(x, np.sin(x-2), color='0.75') #   
                                        #  0  1
plt.plot(x, np.sin(x-3), color='#FFDD44')   #16-  
                                            #(RRGGBB  0  FF)
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3)) # RGB   0  1
plt.plot(x, np.sin(x-5), color='chartreuse') #  (   HTML)
```

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

#     :
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')  #  
plt.plot(x, x+1, '--c') #  
plt.plot(x, x+2, '-.k') #  
plt.plot(x, x+3, ':r')  #   
```

These abbreviations denote RGB(Red/Green/Blue) and CMYK(Cyan/Magenta/Yellow/blacK)

    Axis Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

To display axis in reverse, simply reverse the order of arguments

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

Limits can also be set via plt.axis()
which gets a list of axis limits in form:
[xmin, xmax, ymin, ymax]

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

also axis can be 'equalized'^ so that equal value growth looked similar for both axes

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

plot markers

```python
plt.plot(x, np.sin(x))
plt.title('A Sin Curve') # 

plt.xlabel("x")
plt.ylabel("sin(x)")
```

Use of Legend to mark lines on the plot

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')  #label will set name for the line
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
      xlabel='x', ylabel='sin(x)',
      title='A Simple Plot') # 
```
