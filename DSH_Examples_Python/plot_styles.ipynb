```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np

%matplotlib inline
```

```python
x = np.random.randn(1000)
plt.hist(x)
```

```python
#используем серый фон
ax = plt.axes(facecolor='#E6E6E6')
ax.set_axisbelow(True)

#Рисуем сплошные белые линии сетки
plt.grid(color='w', linestyle='solid')

#скрываем основные линии осей координат
for spine in ax.spines.values():
    spine.set_visible(False)

#Скрываем деления сверху и справа
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()

#осветляяем цвет делений и меток
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')
    
#задаем цвет заливки и границ гистограммы
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666')
```

Для того чтобы сохранить настройки графиков можно воспользоваться rcParams

Каждый раз при запуске matplotlib она описывает конфигурацию среды(rc) содержащую стили по умолчанию для всех создаваемых вами элементов

```python
IPython_default = plt.rcParams.copy()
```

```python
from matplotlib import cycler
colors = cycler('color',
                ['#EE6666', '#3388BB', '#9988DD',
                 '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes', facecolor='#E6E6E6', edgecolor='none',
       axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1, 2, figsize=(11,3))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
```

```python
#Восстанавливаем rcParams
plt.rcParams.update(IPython_default)
```

```python
hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```

```python
with plt.style.context('ggplot'):
    hist_and_lines()
```

```python
with plt.style.context('bmh'):
    hist_and_lines()
```

```python
with plt.style.context('dark_background'):
    hist_and_lines()
```

```python
with plt.style.context('grayscale'):
    hist_and_lines()
```
